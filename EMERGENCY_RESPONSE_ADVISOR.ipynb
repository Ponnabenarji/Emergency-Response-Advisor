```python

DEFAULT_LOCATION = "Chennai"

def callAPI(location):
   
    return {
        "location": location,
        "emergency_type": "Flood",
        "risk_level": "HIGH",
        "weather": "Heavy Rain"
    }

def compare(data):

    if data["risk_level"] == "HIGH":
        return "CRITICAL"
    elif data["risk_level"] == "MEDIUM":
        return "MODERATE"
    return "LOW"

def setState(severity):
    return severity

def guideUser(emergency_type):
    if emergency_type == "Flood":
        return [
            "Stay calm and move to higher ground immediately",
            "Avoid walking or driving through flood water",
            "Turn off electricity if it is safe",
            "Keep emergency contacts accessible"
        ]
    return ["Follow official safety instructions"]

def displayInformation(steps):
    print("\n")
    print(" EMERGENCY RESPONSE ADVISOR ")
    print(" ")
    for step in steps:
        print("-", step)

def run_emergency_advisor(location):
    context = callAPI(location)
    severity = compare(context)
    state = setState(severity)
    steps = guideUser(context["emergency_type"])
    displayInformation(steps)
    return state

final_state = run_emergency_advisor(DEFAULT_LOCATION)
print("\nFinal Emergency State:", final_state)
```
